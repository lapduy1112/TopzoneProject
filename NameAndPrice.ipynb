```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
service_obj=Service(r"D:\chromedriver_win32\chromedriver.exe")
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException
```

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib as plt
```

```python
driver=webdriver.Chrome(service=service_obj)
```

```python
url='https://www.topzone.vn/am-thanh'
```

```python
driver.get(url)
```

```python
elems=driver.find_elements(By.CSS_SELECTOR,'.menu [href]')
title=[elem.text for elem in elems]
title_link=[elem.get_attribute('href') for elem in elems]
```

```python
while True:
    try:
        see_more_button=driver.find_element(By.CSS_SELECTOR,'.view-more')
        see_more_button.click()
        time.sleep(1)
    except ElementNotInteractableException:
        break
    
```

```python
elems_name=driver.find_elements(By.CSS_SELECTOR,'.main-contain h3')
name=[name.text for name in elems_name]
elems_price=driver.find_elements(By.CSS_SELECTOR,'.price')
price=[price.text for price in elems_price]
elems_link=driver.find_elements(By.XPATH,"//a[contains(@class,'main-contain')]")
link=[link.get_attribute("href") for link in elems_link]
oldPrice_list=[]
oldPrice_idx=[]
link
```

```python
for i in range(1,len(name)+1):
    try:
        old_price=driver.find_element(By.XPATH,f"//div[3]/ul/li[{i}]/a/div/p").text
        oldPrice_list.append(old_price)
        oldPrice_idx.append(i)
    except NoSuchElementException:
        print('No such element exception'+str(i))
```

```python
df1=pd.DataFrame(list(zip(name,link,price)),
                columns=['name','link','price'])
df1
```

```python
df1['index']=np.arange(1,len(df1)+1)
```

```python
df2 = pd.DataFrame(list(zip(oldPrice_idx , oldPrice_list)), columns = ['oldPrice_idx','oldPrice_list'])
```

```python
df3=df1.merge(df2,how='left',left_on='index',right_on='oldPrice_idx')
```

```python
df3.to_csv('raw/earphone.csv',index=False)
```
